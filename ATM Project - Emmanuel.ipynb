Design a program to mimic the operation of using an ATM of a bank. This program should perform the following operation:

Display the menu ("Check Balance" , "Deposit", "Withdraw", "EXIT")
Validate the customer pin. If the customer pin is correct, then it should display 
"Welcome! and display the menu, otherwise it can allow customer to try for a maximum of 3 times after which the customer card 
would be blocked.
Check the current balance of the customer (You can assume customer has zero balance at first)
Deposit money to your account.  Display the message "Successful deposit!" and show current balance
Withdraw money from the account and display balance amount. Display the message "Successful withdrawal!" 
if amount to withdraw is possible otherwise "Insufficient balance!".
The program should exit when customer presses the EXIT button

```python

"""Create a class called "ATM_Machine" and define all the attributes of the class"""


class Atm_Machine:
    def __init__(self, first_name, last_name, email_address):
    
        user_pin = int(input("Set Your 4 Digits Access Pin: "))
        self.user_pin = user_pin
        self.fname = first_name
        self.lname = last_name
        self.account_balance = 0
        self.recovery_pin = 9865
        self.email_address = email_address
        
        self.user_pin
        
        """After initializing or authenticating customer's personal details, it takes you to the bank's access page"""
        
        return Atm_Machine.greeting(self)


        """define the various "methods or functions" within the class"""

    def greeting(self):
        print("**************************************************")
        print(f"Welcome To MyBank, {self.fname} {self.lname}.")
        print("**************************************************")
        
        """Opens the page to access your account by aking for your access key or pin"""
        
        login_pin = int(input("Please Enter Your Entry Pin: "))
        
        if login_pin == self.user_pin:
            print("**************************************************")
            print("Login Successful!")
            print("**************************************************")
            
            return Atm_Machine.Main_menu(self)
        
            
            """if user enters a wrong pin, give user three more chances"""
            
        elif login_pin != self.user_pin:
            login_pin_Trial_1 = int(input("WRONG! Please enter the correct pin: "))
            
            if login_pin_Trial_1 == self.user_pin:
                print("**************************************************")
                print("Login Is Successful: Welcome!")
                print("**************************************************")
                
                return Atm_Machine.Main_menu(self)
            
                
            elif login_pin_Trial_1 != self.user_pin:
                login_pin_Trial_2 = int(input("Two More Try: Please enter the correct pin: "))
                
                if login_pin_Trial_2 == self.user_pin:
                    print("**************************************************")
                    print("Login Is Now Successful: Welcome!")
                    print("**************************************************")
                    
                    return Atm_Machine.Main_menu(self)
                    
                    
                elif login_pin_Trial_2 != self.user_pin:
                    login_pin_Trial_3 = int(input("Last Try: Please enter the correct pin: "))
                    
                    if login_pin_Trial_3 == self.user_pin:
                        print("**************************************************")
                        print("Login Finally Successful: Welcome!")
                        print("**************************************************")
                        
                        return Atm_Machine.Main_menu(self)
                        
                        
                        """if user entered wrong pin after all trial, user card should be seized."""

                    else:
                        print("**************************************************")
                        print("Your card is witheld. Please contact the bank's customers service.")
                        return None
                    
                        """Main Menu which allows users to choose from the provided options"""
                    
    def Main_menu(self):
        print ("1). Check Balance")
        print ("2). Deposit")
        print ("3). Withdrawal")
        print ("4). Change Password")
        print ("5). Retrieve Password")
        print ("6). Exit")
            
        menu = int(input("Welcome To MyBank! Please Enter An Option: "))
        
        
        """Checking Account Balance Option"""
        if menu == 1:
            print("**************************************************")
            print(f"Your current balance is: ${self.account_balance}.00")
            print("**************************************************")
            
            print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
            print("If not, 'Click the EXIT option.'")
            
            return Atm_Machine.Main_menu(self)
        
            """Payment or Deposition Option"""    
            
        elif menu == 2:
            credit = int(input("Enter the amount: "))
            self.account_balance += credit
            
            print("**************************************************")
            print("Successful Deposit!")
            print(f"Your New Balance Is: ${self.account_balance}.00")
            print("**************************************************")
            
            print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
            print("If not, 'Click the EXIT option.'")
            
            return Atm_Machine.Main_menu(self)

            
            """Withdrawal Option"""    
            
        elif menu == 3:
            debit = int(input("Enter the amount: "))
            
            """If user tries to withdraw above the available balance, it declines."""
            if debit > self.account_balance:
                print("Insufficient Fund!")
                
                print("**************************************************")
                print(f"Your current balance is: ${self.account_balance}.00")
                print("Please Try A Lesser Amount!")
                print("**************************************************")
                
                return Atm_Machine.Main_menu(self)
            

            else:
                self.account_balance -= debit
                
                print("**************************************************")
                print("Successful withdrawal!")
                print(f"Your New Balance Is: ${self.account_balance}.00")
                print("**************************************************")
            
                print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                print("If not, 'Click the EXIT option.'")

                return Atm_Machine.Main_menu(self)
            
          
        
                """Change of Pin Option"""
            
        elif menu == 4:
            pin_verification = int(input("Enter Your Existing Pin: "))
            
            """Ask of previous pin to allow permission for change of pin"""
            
            if pin_verification != self.user_pin:
                print("**************************************************")
                print("WRONG PIN! Please enter the correct pin.")
                print("**************************************************")
                
                """If users enter the wrong pin, it gives more chances to enter the correct original pin"""
                
                pin_verification = int(input("Enter Your Existing Pin: "))
                if pin_verification != self.user_pin:
                    print("**************************************************")
                    print("WRONG PIN! Please contact the customer service or use the 'Pin Recovery' option from the main menu.")
                    print("**************************************************")

                    return Atm_Machine.Main_menu(self)
                    

                elif pin_verification == self.user_pin:
                    print("**************************************************")
                    print("Existing Pin Has Been Verified!")
                    print("**************************************************")

                new_pin = int(input("Enter Your New Pin: "))
                recheck_new_pin = int(input("Enter Your New Pin Again: "))
                
                """Allows users to validate the new pin to be sure its correct"""

                if new_pin == recheck_new_pin:
                    user_pin = new_pin
                    print("**************************************************")
                    print ("Your pin has been successfully changed!")
                    print("**************************************************")
            
                    print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                    print("If not, 'Click the EXIT option.'")

                    return Atm_Machine.Main_menu(self)
                    

                else:
                    print("**************************************************")
                    print ("Pin confirmation does not match. Try Again!")
                    print("**************************************************")

                    
                    """Gives users more chances to ensure the new pin goes through the double check"""
                    
                    new_pin = int(input("Please Enter Your New Pin: "))
                    recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                    if new_pin == recheck_new_pin:
                        self.user_pin = new_pin
                        print("**************************************************")
                        print ("Your pin has been successfully changed!")
                        print("**************************************************")
            
                        print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                        print("If not, 'Click the EXIT option.'")

                        return Atm_Machine.Main_menu(self)
                    

                    else:
                        print("**************************************************")
                        print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                        return None


            elif pin_verification == self.user_pin:
                print("**************************************************")
                print("Existing Pin Has Been Verified!")
                print("**************************************************")

                new_pin = int(input("Enter Your New Pin: "))
                recheck_new_pin = int(input("Enter Your New Pin Again: "))
                if new_pin == recheck_new_pin:
                    print("**************************************************")
                    self.user_pin = new_pin
                    print ("Your pin has been successfully changed!")
                    print("**************************************************")
            
                    print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                    print("If not, 'Click the EXIT option.'")
                    
                    return Atm_Machine.Main_menu(self)
                    

                else:
                    print("**************************************************")
                    print ("Pin confirmation does not match. Try Again!")
                    print("**************************************************")

                    new_pin = int(input("Please Enter Your New Pin: "))
                    recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                    if new_pin == recheck_new_pin:
                        self.user_pin = new_pin
                        print("**************************************************")
                        print ("Your pin has been successfully changed!")
                        print("**************************************************")
            
                        print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                        print("If not, 'Click the EXIT option.'")

                        return Atm_Machine.Main_menu(self)
                    

                    else:
                        print("**************************************************")
                        print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                        return None
         
        
                        """Retrieve Forgotten Password Option via registered email"""
        
        elif menu == 5:
            customer_email_address = input("Enter Your Registration Email Address: ")
        
            if customer_email_address != self.email_address:
                print("**************************************************")
                print("Wrong email-address. Please try one more time")
                print("**************************************************")

                customer_email_address = input("Enter Your Registration Email Address: ")
                
                """After second trial, the process is aborted if user cannot provide the correct email address"""

                if customer_email_address != self.email_address:
                    print("**************************************************")
                    print("Wrong email-address. Could not verify customer's email-address, please contact our customer service.")
                    return None

                elif customer_email_address == self.email_address:
                    print("**************************************************")
                    print (f"Your recovery pin is: {self.recovery_pin}")
                    print("**************************************************")

                new_user_pin = int(input("Enter the 'recovery pin': "))
                
                """Verifies the recovery pin provided after users have provided the correct email address"""

                if new_user_pin != self.recovery_pin:
                    print("**************************************************")
                    print("Wrong recovery pin. Try Once More!")
                    print("**************************************************")
                    
                    new_user_pin = int(input("Please Enter The Correct Recovery Pin: "))

                    if new_user_pin != self.recovery_pin:
                        print("**************************************************")
                        print("Wrong recovery pin. User Pin Recovery Failed!")
                        return None

                    elif new_user_pin == self.recovery_pin:
                        print("**************************************************")
                        print("Recovery Pin Has Been Verified!")
                        print("**************************************************")

                        new_pin = int(input("Enter Your New Pin: "))
                        recheck_new_pin = int(input("Enter Your New Pin Again: "))
                        
                        """Allows user to set new pin and go through the double check after the recovery pin is verified"""

                    if new_pin == recheck_new_pin:
                        self.user_pin = recheck_new_pin
                        print("**************************************************")
                        print ("Your pin has been successfully changed!")
                        print("**************************************************")

                        print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                        print("If not, 'Click the EXIT option.'")

                        return Atm_Machine.Main_menu(self)


                    else:
                        print("**************************************************")
                        print ("Pin confirmation does not match. Try Again!")
                        print("**************************************************")

                        new_pin = int(input("Please Enter Your New Pin: "))
                        recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                        if new_pin == recheck_new_pin:
                            self.user_pin = new_pin
                            print("**************************************************")
                            print ("Your pin has been successfully changed!")
                            print("**************************************************")
            
                            print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                            print("If not, 'Click the EXIT option.'")

                            return Atm_Machine.Main_menu(self)
                            

                        else:
                            print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                            return None
      

                elif new_user_pin == self.recovery_pin:
                    print("**************************************************")
                    print("Recovery Pin Has Been Verified!")
                    print("**************************************************")

                    new_pin = int(input("Enter Your New Pin: "))
                    recheck_new_pin = int(input("Enter Your New Pin Again: "))

                if new_pin == recheck_new_pin:
                    self.user_pin = recheck_new_pin
                    print("**************************************************")
                    print ("Your pin has been successfully changed!")
                    print("**************************************************")

                    print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                    print("If not, 'Click the EXIT option.'")

                    return Atm_Machine.Main_menu(self)


                else:
                    print("**************************************************")
                    print ("Pin confirmation does not match. Try Again!")
                    print("**************************************************")

                    new_pin = int(input("Please Enter Your New Pin: "))
                    recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                    if new_pin == recheck_new_pin:
                        self.user_pin = new_pin
                        print("**************************************************")
                        print ("Your pin has been successfully changed!")
                        print("**************************************************")

                        print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                        print("If not, 'Click the EXIT option.'")

                        return Atm_Machine.Main_menu(self)


                    else:
                        print("**************************************************")
                        print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                        return None
    
    
            elif customer_email_address == self.email_address:
                print("**************************************************")
                print(f"Your recovery pin is: {self.recovery_pin}")
                print("**************************************************")

                new_user_pin = int(input("Enter the 'recovery pin': "))

                
                if new_user_pin != self.recovery_pin:
                    print("**************************************************")
                    print("Wrong recovery pin. Try Once More!")
                    print("**************************************************")
                    
                    new_user_pin = int(input("Please Enter The Correct Recovery Pin: "))

                    if new_user_pin != self.recovery_pin:
                        print("**************************************************")
                        print("Wrong recovery pin. User Pin Recovery Failed!")
                        return None
                    

                    elif new_user_pin == self.recovery_pin:
                        print("**************************************************")
                        print("Recovery Pin Has Been Verified!")
                        print("**************************************************")

                        new_pin = int(input("Enter Your New Pin: "))
                        recheck_new_pin = int(input("Enter Your New Pin Again: "))

                        if new_pin == recheck_new_pin:
                            self.user_pin = new_pin
                            print("**************************************************")
                            print ("Your pin has been successfully changed!")
                            print("**************************************************")

                            print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                            print("If not, 'Click the EXIT option.'")

                            return Atm_Machine.Main_menu(self)


                    else:
                        print("**************************************************")
                        print ("Pin confirmation does not match. Try Again!")
                        print("**************************************************")

                        new_pin = int(input("Please Enter Your New Pin: "))
                        recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                        if new_pin == recheck_new_pin:
                            self.user_pin = new_pin
                            print("**************************************************")
                            print ("Your pin has been successfully changed!")
                            print("**************************************************")
            
                            print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                            print("If not, 'Click the EXIT option.'")

                            return Atm_Machine.Main_menu(self)
                            

                        else:
                            print("**************************************************")
                            print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                            return None
                
                
                elif new_user_pin == self.recovery_pin:
                    print("**************************************************")
                    print("Recovery Pin Has Been Verified!")
                    print("**************************************************")

                    new_pin = int(input("Enter Your New Pin: "))
                    recheck_new_pin = int(input("Enter Your New Pin Again: "))

                if new_pin == recheck_new_pin:
                    self.user_pin = recheck_new_pin
                    print("**************************************************")
                    print ("Your pin has been successfully changed!")
                    print("**************************************************")

                    print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                    print("If not, 'Click the EXIT option.'")

                    return Atm_Machine.Main_menu(self)


                else:
                    print("**************************************************")
                    print ("Pin confirmation does not match. Try Again!")
                    print("**************************************************")

                    new_pin = int(input("Please Enter Your New Pin: "))
                    recheck_new_pin = int(input("Please Enter Your New Pin Again: "))

                    if new_pin == recheck_new_pin:
                        self.user_pin = new_pin
                        print("**************************************************")
                        print ("Your pin has been successfully changed!")
                        print("**************************************************")

                        print("Do you want to perform another operation? If yes, 'Choose from the below options.'")
                        print("If not, 'Click the EXIT option.'")

                        return Atm_Machine.Main_menu(self)


                    else:
                        print("**************************************************")
                        print("Existing Pin Verification Failed. Pin Change Cannot Be Completed!")
                        return None              
        
                        """Once all activities is finished, users can exit."""
            
        elif menu == 6:
            print("**************************************************")
            print("Thank You For Banking With MyBank. Have A Lovely Day!")
            return None
            

            """if user entered an invalid number from the available options"""
            
            
        else:
            print("**************************************************")
            print("Please Enter A Valid Number From The Given Options:")
            print("**************************************************")
            
            """When an invalid option is entered, Return to the main menu to re-enter a valid option"""
            
            return Atm_Machine.Main_menu(self)
        
            
```
